```python
#Tabla Estados
mycursor.execute("CREATE TABLE IF NOT EXISTS estados (stateID int AUTO_INCREMENT, state varchar(255), primary key (stateID))")
```

```python
#Tabla Clientes
mycursor.execute("CREATE TABLE IF NOT EXISTS clientes (clientID int, name varchar(255), primary key (clientID))")
```

```python
#Tabla Categoria
mycursor.execute("CREATE TABLE IF NOT EXISTS categorias (categoryID int AUTO_INCREMENT, category varchar(255), primary key (categoryID))")
```

```python
#Tabla Local
mycursor.execute("CREATE TABLE IF NOT EXISTS locales (storeID int, stateID int not null,  name varchar(255), address varchar(255), latitude decimal(9,6), longitude decimal(9,6), delivery varchar(255), primary key (storeID), FOREIGN KEY (stateID) REFERENCES estados(stateID))")
```

```python
#Tabla intermedia Categoria_Local
mycursor.execute("CREATE TABLE IF NOT EXISTS categoria_local (categoryID int not null, storeID int not null, FOREIGN KEY (categoryID) REFERENCES categorias(categoryID), FOREIGN KEY (storeID) REFERENCES locales(storeID) )")
```

```python
#Tabla reviews
mycursor.execute("CREATE TABLE IF NOT EXISTS reviews (reviewID int AUTO_INCREMENT, clientID int not null, storeID int not null, date datetime, rating int, text varchar(255),primary key (reviewID), FOREIGN KEY (clientID) REFERENCES clientes(clientID), FOREIGN KEY (storeID) REFERENCES locales(storeID))")
```

```python
#Tabla Comentarios
mycursor.execute("CREATE TABLE IF NOT EXISTS comentarios (commentID int AUTO_INCREMENT, reviewID int not null, text varchar(255), date datetime, primary key (commentID), FOREIGN KEY (reviewID) REFERENCES reviews(reviewID))")
```
